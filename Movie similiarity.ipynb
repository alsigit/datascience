```python
import pandas as pd
```

# import dataset

```python
kolom_rating = ['user_id', 'movie_id', 'rating']
rating = pd.read_csv('dataset/u.data', sep = '\t', names = kolom_rating, usecols = range(3), encoding = "ISO-8859-1")
```

```python
kolom_film = ['movie_id', 'title']
film = pd.read_csv('dataset/u.item', sep = '|', names = kolom_film, usecols = range(2), encoding = "ISO-8859-1")
```

```python
rating = pd.merge(film, rating)
```

```python
rating.head()
```

# data preprocessing

```python
movieRating= rating.pivot_table(index = ['user_id'], columns = ['title'], values = 'rating')
```

```python
movieRating.head()
```

```python
starWarsRating = movieRating['Toy Story (1995)']
```

```python
starWarsRating.head()
```

```python
similiarMovie = movieRating.corrwith(starWarsRating)
```

```python
similiarMovie = similiarMovie.dropna()
```

```python
similiarMovie.head()
```

```python
df = pd.DataFrame(similiarMovie)
```

```python
df.head()
```

```python
similiarMovie.sort_values(ascending = False).head()
```

```python
import numpy as np
```

```python
movieStats = rating.groupby('title').agg({'rating' : [np.size, np.mean]})
```

```python
movieStats.head()
```

```python
popularMovie = movieStats['rating']['size'] >=100
```

```python
popularMovie
```

```python
movieStats[popularMovie].sort_values([('rating', 'mean')],ascending=False)[:15]
```

```python
df = movieStats[popularMovie].join(pd.DataFrame(similiarMovie, columns = ['similiarity']))
```

```python
df.head()
```
